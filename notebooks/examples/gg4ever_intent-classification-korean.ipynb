### Huggingface

https://huggingface.co/gg4ever/intent-classification-korean


### Training data

fine-tuned for 'klue/roberta-base' used data : 'kor_3i4k'


### Label info

- 0: "fragment",
- 1: "statement",
- 2: "question",
- 3: "command",
- 4: "rhetorical question",
- 5: "rhetorical command",
- 6: "intonation-dependent utterance"


```python
from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    TextClassificationPipeline,
)

model_name = "gg4ever/intent-classifcation-korean"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)
classifier = TextClassificationPipeline(model=model, tokenizer=tokenizer)
```

```python
res = classifier("what is the weather like today?")
res
```

```python
res = classifier("오늘 날씨가 어때?")
res
```
